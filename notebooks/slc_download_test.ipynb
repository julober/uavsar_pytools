```python
import os
from uavsar_pytools.download.download_slcs import get_uavsar_slcs, download_uavsar_slcs
```

```python
# Create a temporary directory for our test downloads
test_dir = './uavsar_test_downloads'
os.makedirs(test_dir, exist_ok=True)
test_dir = os.path.abspath(test_dir)
print(f"Created test directory at {test_dir}")
```

```python
# Query links for a narrow time window
test_links_dict = get_uavsar_slcs(
    flight_name='lowman',
    start_date='2020-02-01',
    end_date='2020-02-28',
    tag=['BC'],
    seg=['s1', 's2'],
    getann=True,
    getdop=True,
    getllh=True,
    getlkv=True
)

# Extract the dictionary key (should be 'grmesa_03913' or similar)
test_key = list(test_links_dict.keys())[0]
generated_urls = test_links_dict[test_key]

print(f"Generated {len(generated_urls)} URLs for {test_key}.")
print("\nFirst 3 URLs found:")
for url in generated_urls[:]:
    print(url)
```

```python
# Filter the list to ONLY include an .ann file to save bandwidth
ann_files_only = [url for url in generated_urls if url.endswith('.ann')]

if not ann_files_only:
    print("No .ann files found. Make sure getann=True was passed in Cell 3.")
else:
    # We will just pass the very first .ann file to the download function
    test_download = ann_files_only
    
    print(f"Testing download pipeline with: {test_download}")
    
    # Run the download
    download_uavsar_slcs(test_download, out_dir=test_dir)
    
    # Verify it exists
    downloaded_path = os.path.join(test_dir, test_download[0])
    if os.path.exists(downloaded_path):
        print(f"Found file verified on disk at {downloaded_path}")
    else:
        print("Download failed or file not found.")
```
